```python
icon_name = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
train = []
target = []
for i in range(1,6):
    file_path = 'data_batch_%d'%(i)
    with open('data_batch_1','rb') as fp:
        temp = pickle.load(fp,encoding='ISO-8859-1')
        data = temp['data']
        labels = temp['labels']
    train.append(data)
    target.append(labels)
```

```python
samples = np.concatenate(train)
labels = np.concatenate(target)
```

```python
display(samples.shape,labels.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```

```python
train = samples[:2500]
target = labels[:2500]
```

```python
pca = PCA(n_components=0.9)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(train,target,test_size=0.2,random_state=4)
```

```python
# 算法评分
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
svc = SVC(kernel='rbf',C=10)
knn = KNeighborsClassifier()
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(svc,X_train,y_train)
```

```python
cross_val_score(knn,X_train,y_train)
```

```python
X_train_pca = pca.fit_transform(X_train)
knn.fit(X_train_pca,y_train)
```

```python
X_test_pca = pca.transform(X_test)
y_ = knn.predict(X_test_pca)
```

```python
knn.score(X_test_pca,y_test)
```
